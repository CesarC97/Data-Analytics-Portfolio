## Data Cleaning

```python
# Dropping LocationAbbrev and LocationDesc
smoking_data = smoking_data.drop(['LocationAbbrev', 'LocationDesc'], axis = 1)

# Remove total categories to avoid overcounting
totals_to_remove = ['Total Cigars', 'Total Loose Tobacco', 'Total Combustible Tobacco']
smoking_data = smoking_data[~smoking_data['Submeasure'].isin(totals_to_remove)].reset_index(drop=True)

# Get population as millions
smoking_data['Pop_Per_Million'] = smoking_data['Population'] / 1000000
smoking_data['Domestic_Per_Million']  = smoking_data['Domestic'] / smoking_data['Pop_Per_Million']
smoking_data['Imports_Per_Million'] = smoking_data['Imports'] / smoking_data['Pop_Per_Million']
smoking_data['Total_Per_Million'] = smoking_data['Domestic_Per_Million'] + smoking_data['Imports_Per_Million']

# Add new columns and drop original ones
smoking_data.drop(columns = ['Pop_Per_Million', 'Domestic', 'Imports', 'Total', 
                             'Domestic Per Capita', 'Imports Per Capita', 'Total Per Capita'],
                  inplace = True)

# Keep only 'Cigarette Equivalents' for Roll-Your-Own and Pipe Tobacco
mask = (
    ((smoking_data['Submeasure'] == 'Roll-Your-Own Tobacco') | 
     (smoking_data['Submeasure'] == 'Pipe Tobacco')) & 
    (smoking_data['Data Value Unit'] != 'Cigarette Equivalents')
)

# Drop those rows (i.e., the ones using 'Pounds' for these submeasures)
smoking_data = smoking_data[~mask].reset_index(drop=True)

smoking_data.head(7)
```

The data cleaning made the data tidier and ready for exploratory data analysis.
Such methods include:

- As mentioned before, there are two unary variables, ``LocationAbbrev`` and ``LocationDesc`` that were deleted, as the dataset covers US National data
- The data needs different categories for the ``Submeasure`` variable to avoid overcounting issues; values such as *Total Cigars*, *Total Loose Tobacco*, and *Total Combustible Tobacco* were deleted
- Rows where all *Per Capita* values were approximately equal to zero were converted into *Per Million* which gives easier-to-read figures, and ``Domestic``, ``Imports``, and ``Total`` columns were deleted to focus more on the smoking behavior

## Exploratory Data Analysis

### Trend of Cigarette Consumption (2000 - 2023)

```python
# Filter for cigarettes
cigs = smoking_data[smoking_data['Submeasure'] == 'Cigarette Removals']

plt.figure(figsize=(10, 6))
sns.lineplot(data=cigs, x='Year', y='Total_Per_Million', marker='o')
plt.title("Per Million Cigarette Consumption (2000–2023)")
plt.ylabel("Cigarettes per Million")
plt.grid(True)
plt.tight_layout()
plt.show()
```

This line plot shows the annual per capita cigarette consumption per million people in the United States from 2000 to 2023. The data is adjusted for population growth, providing a normalized view of consumption trends over time. Based on this plot, there are some interesting insights about cigarette consumption:

- Cigarette consumption has shown a steady and significant decline over the 23-year period
- In 2000, Americans consumed over 2.1 billion cigarettes per million people
- By 2023, this dropped to below 700 million cigarettes per million — roughly a 67% decrease
- A sharper drop is observed around 2009 - 2010, the decline continues, although the slope becomes slightly more gradual in the later years
- From around 2013 to 2016, the decline slows, suggesting a potential stabilization in smoking reduction efforts
- Around 2020, there’s a small upward bump, potentially linked to behavioral shifts during the COVID-19 pandemic (stress-related consumption, disruptions in cessation programs), but the downward trend resumes after 2021

This trend strongly supports the conclusion that adult cigarette smoking has significantly declined in the U.S. over the past two decades. Factors contributing to this trend likely include:

- Tobacco regulation and taxation

- Public health campaigns

- Cultural shifts in smoking norms

- Increased use of nicotine alternatives (e.g., e-cigarettes, vaping, snus)

While this is a positive public health indicator, further analysis should explore whether other forms of tobacco (e.g., cigars, smokeless tobacco) have replaced cigarettes — especially among younger demographics or specific subgroups.

### Combustible vs Noncombustible Products Over Time

```python
# Tag categories
def classify_type(sub):
    if sub in ['Snuff', 'Chewing Tobacco']:
        return 'Noncombustible'
    return 'Combustible'

smoking_data['Product_Type'] = smoking_data['Submeasure'].apply(classify_type)

# Group by Year and Product_Type
summary = smoking_data.groupby(['Year', 'Product_Type'])['Total_Per_Million'].sum().reset_index()

# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=summary, x='Year', y='Total_Per_Million', hue='Product_Type', marker='o')
plt.title("Combustible vs Noncombustible Tobacco Use (Per Million)")
plt.ylabel("Total Units per Million People")
plt.grid(True)
plt.tight_layout()
plt.show()
```

This visualization compares the per capita tobacco use per million people in the U.S. from 2000 to 2023, separated into two broad product categories:

- Combustible tobacco (e.g., cigarettes, cigars, roll-your-own, pipe tobacco)

- Noncombustible tobacco (e.g., chewing tobacco, snuff)

There are some interesting observations about this plot:

- Throughout the entire time period, combustible tobacco use far outweighs noncombustible use — by several orders of magnitude

- In 2000, usage was over 2.1 billion units per million people, steadily declining over time

- In contrast, noncombustible tobacco stays consistently low and flat, barely visible on the same scale

- Mirroring the cigarette-specific chart earlier, combustible use has dropped dramatically — by more than 65% over the 23 years

- This supports the broader public health narrative: smoking behavior is declining nationwide

- Unlike combustibles, noncombustible tobacco use has remained relatively flat
- The chart suggests that as combustible use drops, users are not widely switching to smokeless tobacco (like chewing tobacco or snuff)

## Hypothesis Testing

- Null Hypothesis

 $H_0$: There is no significant change in per capita cigarette consumption over time.

- Alternative Hypothesis

 $H_1$: Per capita cigarette consumption has decreased over time.

### Linear Regression (preferred for time trends)

```python
import statsmodels.api as sm

# Filter for cigarette data only
cigs = smoking_data[smoking_data['Submeasure'] == 'Cigarette Removals']
X = cigs['Year']
y = cigs['Total_Per_Million']

# Add constant for intercept
X = sm.add_constant(X)

# Fit regression model
model = sm.OLS(y, X).fit()
print(model.summary())
```

A simple linear regression was conducted to assess the relationship between year and per capita cigarette consumption (measured in total units per million people) in the U.S. from 2000 to 2023. The results showed a significant negative trend:

- Each year, cigarette consumption per million people decreased by approximately 60 million units (β = -60,040,000, p < 0.001).

- The model explained 97.6% of the variation in cigarette use across years (R² = 0.976), indicating a very strong fit.

This provides statistical evidence that adult cigarette consumption in the U.S. has significantly declined over the past two decades, supporting public health surveillance findings and trends observed in the exploratory data analysis.

### Two-Sample T-Test 

```python
from scipy.stats import ttest_ind

# Group by time period
early = cigs[cigs['Year'] <= 2005]['Total_Per_Million']
late = cigs[cigs['Year'] >= 2018]['Total_Per_Million']

# Perform one-tailed t-test (alternative = 'less')
t_stat, p_val = ttest_ind(early, late, alternative='less')
print("p-value:", p_val)
```

A two-sample t-test was performed comparing per capita cigarette consumption between the early 2000s (2000–2005) and recent years (2018–2023). The results showed a significant decrease in consumption over time (p < 0.001), supporting the hypothesis that adult smoking in the U.S. has declined over the past two decades.
